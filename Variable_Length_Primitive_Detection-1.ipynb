```python
#Calculation of entropy function for a list of values

def entropy(arr):
    s=0
    for i in arr:
        s+=-i/sum(arr)*math.log2(i/sum(arr))
    return s
```

### Building the Tree

Through tree object, we are observing an increase in entropy as we move add characters to substring

```python
class tree(object):
    def __init__(self, char: str):
        self.char = char
        self.children = []
        self.counter = 1
        self.end=0
        self.freq=0
```

```python
#Function to add elements to the tree

def add(root, word: str,variation,frequency):
    node = root
    for char in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.char == char:
              #child.counter += 1
              node = child
              found_in_child = True
              break
        # We did not find it so add a new chlid
        if not found_in_child:
          new_node = tree(char)
          node.children.append(new_node)
          # And then point node to the new child
          node = new_node
    node.counter=variation
    node.freq=frequency
```

```python
#Function which add all elements of dictionary to the tree 

def primitives(d):
  root = tree('*')
  for i in d.keys():
    add(root,i,d[i][0],d[i][1])
  return root

root=primitives(d)
root.counter=-1  
```

```python
#Extracting threshold of frequency for fixed length primitive

def threshold(g):                            
  l=sorted(list(g.values()))
  m=[]
  t=-1
  val=0
  prev=0
  prev=0
  diff=0
  for i in range(1,len(l)-1):
    if l[i]-l[i-1]>diff:
      diff=l[i]-l[i-1]
      prev=val
      val=l[i-1]
  for i in g.keys():
    if g[i]>=prev:
      m.append(i)
  return set(m)
```

### Further Filtering

This keeps the primitives which are foundational (which cannot be expressed as combination of others)

```python
class subtree(object):
    def __init__(self, char: str):
        self.char = char
        self.children = []
        self.counter = 0
        self.end=0
```

```python
def subadd(root, word: str):
    if word=='':
      return
    node = root
    for char in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.char == char:
              #child.counter += 1
              node = child
              found_in_child = True
              break
        # We did not find it so add a new chlid
        if not found_in_child:
          new_node = tree(char)
          node.children.append(new_node)
          # And then point node to the new child
          node = new_node
    node.end=1
    node.counter+=1
```

```python
def subprint_tree(root,s,final,length):
  s+=root.char
  if root.children!=[] and root.end==1 and root.counter>1:
    final.append(s[1:])
  if root.children==[]:
    final.append(s[1:])
    s=''
    length=0
    return 0 
  for child in root.children:
    subprint_tree(child,s,final,length+1)
```

```python
def search(root,word):
  if word=='':
    return 
  node = root
  c=0
  for char in word:
    c+=1
    found_in_child = False
    for child in node.children:
      if child.char == char:
        node = child
        found_in_child = True
        word=word[1:]
        break
  if found_in_child==False:
    return
  if len(word)==0 and node.children==[]:
    return '*'
```

```python
def add(root,word):
  c=0
  for i in range(len(word)):
    if search(root,word[i:])=='*' or search(root,word[:i])=='*':
      if search(root,word[i:])=='*':
          word=word[:i]
          add(root,word)
      elif search(root,word[:i])=='*':
          word=word[i:]
          add(root,word)
      c=1
      break 
  if c!=1:
    subadd(root,word)
```
