```python
class tree(object):
    def __init__(self, char: str):
        self.char = char
        self.children = []
        self.counter = 1
        self.end=0
        self.freq=0
```

```python
def add(root, word: str,variation,frequency):
    node = root
    for char in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.char == char:
              #child.counter += 1
              node = child
              found_in_child = True
              break
        # We did not find it so add a new chlid
        if not found_in_child:
          new_node = tree(char)
          node.children.append(new_node)
          # And then point node to the new child
          node = new_node
    node.counter=variation
    node.freq=frequency
```

```python
def ent_string(word,string):
  f={}
  b={}
  for i in range(len(string)):
    if string[i:i+len(word)]==word:
      if i>0:
        try:
          f[string[i-1]]+=1
        except:
          f[string[i-1]]=1
      if i+len(word)<len(string):
        try:
          b[string[i+len(word)]]+=1
        except:
          b[string[i+len(word)]]=1
  return min(entropy(list(f.values())),entropy(list(b.values())))
```

```python
def residue(k,string):
    x=0
    d={}
    e={}
    k=sorted(k,reverse=True)
    i=0
    c=0
    while i<len(string):
        t=0
        for j in k:
            if string[i:len(j)+i]==j:
                t=1
                i=i+len(j)
                break
        if t==0:
            i=i+1
            c=c+1
    
    return(c/len(string))
```

```python
def guide(k,string):
    x=0
    d={}
    e={}
    k=sorted(k,reverse=True)
    i=0
    c=0
    while i<len(string):
        t=0
        for j in k:
            if string[i:len(j)+i]==j:
                t=1
                try:
                    d[len(j)]+=1
                except:
                    d[len(j)]=1
                try:
                    e[j]+=1
                except:
                    e[j]=1
                i=i+len(j)
                break
        if t==0:
            i=i+1
            c=c+1
    
    
    for i in d.keys():
        x+=d[i]*i
    
    try:
        return(sum(d.values())/x+2*math.ceil(math.log2(len(k)))*sum(e.values())/len(string))
    except:
        return -1
```

```python
#Split the existing elements based on 3 conditions:
#Entropy of either child is greater than its parents
#Frequency of both children is greater than the parent
#Difference in entropy of 2 children should be maximum

def split_primitives(word,fre,string):

  if ent_string(word,string)==0 and fre[word]>1:
    return -1,-1

  diff=0
  
  diff_max=0
  index=-1
  for i in range(1,len(string)-1):
    if ent_string(word[:i],string)>ent_string(word,string) or ent_string(word[i:],string)>ent_string(word,string) and min(ent_string(word[:i],string),ent_string(word[i:],string))!=0:
      if fre[word[:i]]>fre[word] and fre[word[i:]]>fre[word]:
        diff=abs(ent_string(word[:i],string)-ent_string(word[i:],string))
        if diff>diff_max:
          index=i
          diff_max=diff
  if index!=-1:
    i=index
    return word[:i],word[i:]
  else:
    return -1,-1
```

###Building Tree for extracting maximum length suffix and prefix

```python
class st(object):
    def __init__(self, char: str):
        self.char = char
        self.children = []
        self.counter = 1
        self.end=0
        self.freq=0
```

```python
def st_add(root, word):
    node = root
    for char in word:
        found_in_child = False
        # Search for the character in the children of the present `node`
        for child in node.children:
            if child.char == char:
              #child.counter += 1
              node = child
              found_in_child = True
              break
        # We did not find it so add a new chlid
        if not found_in_child:
          new_node = st(char)
          node.children.append(new_node)
          # And then point node to the new child
          node = new_node
    node.end=1
    node.freq=fre
```

```python
def st_search(root,word):
  if word=='':
    return 
  node = root
  c=0
  for char in word:
    c+=1
    found_in_child = False
    for child in node.children:
      if child.char == char:
        node = child
        found_in_child = True
        word=word[1:]
        if word=='':
          return node.end
        break
  if found_in_child==False:
    return -1
  if len(word)==0 and node.children==[]:
    return node.end
```

###Combine Primitives Section

```python
#Finding Suffix of a Primitive for Combination

def suffix_replacement(k,word,fre,string):
  root=st('*')
  for i in k:
    st_add(root,i)

  jmax=0
  for i in k:
    if len(i)>jmax:
      jmax=len(i)

  mat1=np.zeros(len(k))

  ind={}
  for i in range(len(k)):
    ind[k[i]]=i


  for t in range(len(string)):
      if string[t:len(word)+t]==word:
        j=jmax
        while t+j+len(word)>t+len(word):
          b=st_search(root,string[t+len(word):t+j+len(word)])
          if b==1:
            mat1[ind[string[t+len(word):t+j+len(word)]]]+=1
            break
          j-=1

  rep=[]
  if sum(mat1)==fre[word]:
    for i in range(len(mat1)):
      if mat1[i]!=0 and word+k[i] in fre.keys() and fre[word+k[i]]>1:
        rep.append(word+k[i])




  tr=m.copy()
  tr.remove(word)
  tr+=rep
  if residue(m,string)==residue(tr,string) and guide(m,string)>guide(tr,string) and rep!=[]:
    return rep,guide(m,string)-guide(tr,string)

  return [],0
```

```python
#Finding Prefix of a Primitive for Combination

def prefix_replacement(k,word,fre,string):
  root=st('*')
  for i in k:
    st_add(root,i)

  fmax=0
  for i in k:
    if len(i)>fmax:
      fmax=len(i)

  mat1=np.zeros(len(k))

  ind={}
  for i in range(len(k)):
    ind[k[i]]=i


  for t in range(len(string)):
      if string[t:len(word)+t]==word:
        f=fmax
        while t-f<t:
          b=st_search(root,string[t-f:t])
          if b==1:
            mat1[ind[string[t-f:t]]]+=1
            break
          f-=1

  rep=[]
  if sum(mat1)==fre[word]:
    for i in range(len(mat1)):
      if mat1[i]!=0 and k[i]+word in fre.keys() and fre[k[i]+word]>1:
        rep.append(k[i]+word)

  tr=m.copy()
  tr.remove(word)
  tr+=rep
  if residue(m,string)==residue(tr,string) and guide(m,string)>guide(tr,string) and rep!=[]:
    return rep,guide(m,string)-guide(tr,string)

  return [],0
```

###Implementing Max and Min Prefix for Combining Primitives

```python
def combine_primitives(m,fre,string):
  jo=[]
  c=0
  for i in set(m):
    c+=1
    t1,res1=suffix_replacement(m,i,fre,string)
    t2,res2=prefix_replacement(m,i,fre,string)
    if t1==[] and t2==[]:
        jo+=[i] 
    elif res1>res2:
      jo+=t1
    elif res2>res1:
      jo+=t2
  return jo

jo=combine_primitives(m,fre,string)
```
